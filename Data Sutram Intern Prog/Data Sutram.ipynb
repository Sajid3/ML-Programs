#     Weighted HeatMap

```python
#IMPORTS
import pandas as pd
import folium
from folium import plugins
from folium.plugins import HeatMap
```

```python
#Data preprocessing 
df0 = pd.read_excel('heatmap_buy_20.xlsx',  skiprows = range(1),skipfooter = 0)
df0['weight']=df0['weight']/max(df0['weight'])
df0
```

```python
kmap = folium.Map(location=[22.5726, 88.3639],zoom_start = 11) 

df_acc = df0
df0['latitude'] = df0['latitude'].astype(float)
df0['longitude'] = df0['longitude'].astype(float)

df1 = df0[['latitude', 'longitude']]

heat_data = [[row['latitude'],row['longitude']] for index, row in df1.iterrows()]


for index, row in df1.iterrows():
    folium.CircleMarker([row['latitude'], row['longitude']],radius=3,popup=row,fill_color="red").add_to(kmap)

# Plot 
kmap.add_child(plugins.HeatMap(heat_data,blur=10,gradient={.2: 'blue', .65: 'lime',.8:'yellow' , 1: 'red'} ,radius=110))

# Display 
kmap
```
