# Q.2a.Find out the top 5 countries from Asia who have contributed maximum total immigration into Canada over the entire period of 1980 to 2013

```python
dfcan=df_canada
dfcan=dfcan[dfcan['AreaName'].str.match('Asia')]
dfcan
```

```python
dfcan.drop(['Type','Coverage','DevName','DEV','RegName','AreaName','AREA','REG'], axis=1, inplace=True)
dfcan.rename(columns={'OdName':'Country', 'AreaName':'Continent'}, inplace=True)

dfcan['Total']=dfcan.sum(axis=1)
dfcan
```

```python
tdfcan=dfcan[['Country','Total']]
#sortedtdfcan=tdfcan.sort_values('Total',ascending=False)
#x=(((tdfcan.iloc[:,-1])/max(tdfcan.iloc[:,-1])))
x=(tdfcan.iloc[:,-1])/1000
#  finaldf=tdfcan[['Country',]]

nrmtdfcan=pd.DataFrame(tdfcan["Country"])
#nrmtdfcan


finaldf=nrmtdfcan.join(x).sort_values('Total',ascending=False).head(5)
finaldf
```

```python

# Plot the figure.
plt.figure(figsize=(25, 90))
ax = finaldf.plot(kind='barh')

ax.set_title('Top Immigration')
ax.set_xlabel('Total')
ax.set_ylabel('Country')
ax.set_yticklabels(finaldf["Country"])
ax.set_xlim(0, 700) # expand xlim to make labels easier to read

rects = ax.patches

# For each bar: Place a label
for rect in rects:
    # Get X and Y placement of label from rect.
    x_value = rect.get_width()
    y_value = rect.get_y() + rect.get_height() / 2

    # Number of points between bar and label. Change to your liking.
    space = 5
    # Vertical alignment for positive values
    ha = 'left'

    # If value of bar is negative: Place label left of bar
    if x_value < 0:
        # Invert space to place label to the left
        space *= -1
        # Horizontally align label at right
        ha = 'right'

    # Use X value as label and format number with one decimal place
    label = "{:.3f}".format(x_value)

    # Create annotation
    plt.annotate(
        label,                      # Use `label` as label
        (x_value, y_value),         # Place label at end of the bar
        xytext=(space, 1),          # Horizontally shift label by `space`
        textcoords="offset points", # Interpret `xytext` as offset in points
        va='center',                # Vertically center label
        ha=ha)                      # Horizontally align label differently for
                                    # positive and negative values.

#plt.savefig("image.png")
plt.show()
```

# 2.a.Plot the immigration trend from the above 5 countries over the years from 1980 thru 2013

```python
#dfcan[finaldf.index.isin(dfcan.index)]

dfcan=dfcan.loc[dfcan.index & finaldf.index]
dfcan
```

```python
dfcan.set_index('Country',inplace=True)
dfcan
```

```python
years = list(map(int, range(1980, 2014)))

df_Immigration = dfcan.loc[['India', 'China','Philippines','Iran','Pakistan'], years]
df_Immigration.transpose().plot(kind = 'line',linewidth=1,marker='*')
plt.title('Immigration Trends of 5 Countries')
plt.ylabel('Number of immigrants')
plt.xlabel('Years')

plt.show()
```
